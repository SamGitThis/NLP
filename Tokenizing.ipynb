```python
mystr = "Hello I am Sam. This is a sample paragraph. Real Madrid vs Man City ended 1-1. It was a very good game."
```

```python
from nltk.tokenize import word_tokenize
mystr_tokenize = word_tokenize(mystr)
mystr_tokenize
```

```python
len(mystr_tokenize)
```

```python
from nltk.probability import FreqDist
fdist = FreqDist()

for word in mystr_tokenize:
    fdist[word.lower()] += 1
fdist
```

```python
fdist['.'] #number of fullstops
```

```python
print(len(mystr))
len((mystr_tokenize))
```

```python
len(fdist)
```

```python
from nltk.util import bigrams, trigrams, ngrams
```

```python
strg = "The world is beautiful right"
tokstrg = word_tokenize(strg)
tokstrg
```

```python
tokstrg_bigram = list((nltk.bigrams(tokstrg)))
tokstrg_bigram
```

```python
tokstrg_trigram = list((nltk.trigrams(tokstrg)))
tokstrg_trigram
```

```python
tokstrg_ngram = list((nltk.ngrams(tokstrg, 4)))
print(tokstrg_ngram)

tokstrg_ngram = list((nltk.ngrams(tokstrg, 5)))
print(tokstrg_ngram)
```
